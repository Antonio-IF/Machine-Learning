# ¿Qué es el Machine Learning? 

###### [ESP] Existe un tercer tipo de aprendizaje en machine learning, menos común, conocido como aprendizaje por refuerzo (Reinforcement Learning). En este enfoque, el modelo —denominado agente— toma decisiones o realiza acciones dentro de un entorno, buscando maximizar una recompensa acumulada. Si el agente se comporta como se espera, recibe un refuerzo positivo (recompensa); en caso contrario, recibe un refuerzo negativo (castigo), lo cual le permite ajustar su comportamiento a lo largo del tiempo para mejorar su desempeño en la tarea objetivo.


- El Machine Learning (Aprendizaje automático) es una rama de la Inteligencia Artificial (IA) centrada en entrenar a computadoras y máquinas para imitar el modo en que aprenden los humanos, realizar tareas de forma autónoma y mejorar su rendimiento y precisión a través de la experiencia y la exposición a más datos.
